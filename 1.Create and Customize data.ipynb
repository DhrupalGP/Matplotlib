```python
# import needed module
import matplotlib.pyplot as plt
```

```python
# ages of developer
ages_x = list(range(25,36))
ages_x
```

```python
# Media salary of all developers
sal_y = [38496,42000,46742,49320,53200,56000,62316,64928,67317,68748,73752]
sal_y
```

```python
# Medain Python developer salary
py_dev_y = [45372, 48876, 53850, 57287, 63016,65998, 70003, 70000, 71496, 75370, 83640]
py_dev_y
```

```python
# Median JavaScript Developer Salaries
js_dev_y = [37810, 43515, 46823, 49293, 53437,56373, 62375, 66674, 68745, 68746, 74583]
```

```python
# plot line graph

plt.style.use(plt.style.available[25])

plt.plot(ages_x, sal_y, color="k", linestyle="-", marker="<",label="All dev")
plt.plot(ages_x, py_dev_y, color="b", marker="*", label="Python dev")
plt.plot(ages_x,js_dev_y, color="r", linestyle="-", marker="p", label="JS dev")

plt.title("Median Salary(USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary(USD)")
plt.legend()

# plt.xkcd()
plt.grid(True)
plt.tight_layout()
plt.savefig("plot.png")
plt.show()
```
