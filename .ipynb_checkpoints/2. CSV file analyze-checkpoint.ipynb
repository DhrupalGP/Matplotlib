```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
age = data['Age']
dev_sal = data['All_Devs']
py_sal = data['Python']
js_sal = data['JavaScript']
age.head(),dev_sal.head(),py_sal.head(),js_sal.head()
```

```python
plt.style.use(plt.style.available[10])

plt.plot(age,dev_sal,label="All Devel",color="blue",marker="o")
plt.plot(age,py_sal,label="Python",color="red", marker="H")
# plt.plot(age,js_sal,label="Javascript",color="black", marker="^")

plt.legend()

overall_median = 60000

plt.fill_between(age,py_sal,overall_median,
                 where=(py_sal > overall_median),
                 interpolate=True,
                 alpha=0.25)

plt.fill_between(age,py_sal,overall_median,
                 where=(py_sal <= overall_median),
                 interpolate=True,
                 alpha=0.25,color="red")

plt.title("Age VS. Salaries of developers in(USD)")
plt.xlabel("Age of developers")
plt.ylabel("Salaries of developers in(USD)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.style.use(plt.style.available[10])

plt.plot(age,dev_sal,label="All Devel",color="blue",marker="o")
plt.plot(age,py_sal,label="Python",color="red", marker="H")

plt.fill_between(age,py_sal,dev_sal,
                 where=(py_sal > dev_sal),
                 interpolate=True,
                 alpha=0.25,label="Above avg.")

plt.fill_between(age,py_sal,dev_sal,
                 where=(py_sal <= dev_sal),
                 interpolate=True,
                 alpha=0.25,color="red",label="Below avg.")
plt.legend()

plt.title("Age VS. Salaries of developers in(USD)")
plt.xlabel("Age of developers")
plt.ylabel("Salaries of developers in(USD)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
