```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
```

```python
ages = [18,19,21,25,26,26,30,32,38,45,55]
bins = [10,20,30,40,50,60]
```

```python
#bins are used for range in histogram
plt.hist(ages,bins=bins,edgecolor="black")

plt.title("Histogram")
plt.tight_layout()
plt.grid()
plt.show()
```
